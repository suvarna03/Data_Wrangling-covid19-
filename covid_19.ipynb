## 4. Filter the dataframe with only this columns
###   ['continent','location','date','total_cases','total_deaths','gdp_per_capita','
###   human_development_index'] and update the data frame.

```python
data=data.loc[:,['continent','location','date','total_cases','total_deaths','gdp_per_capita','human_development_index']]
```

```python
data.shape
```

## 5. Data Cleaning

### a. Remove all duplicates observations

```python
df = data.drop_duplicates()
```

```python
df.shape
```

###  b. Find missing values in all columns

```python
df.isna().sum()
```

### c. Remove all observations where continent column value is missing
 ###       Tip : using subset parameter in dropna

```python
df.dropna(axis = 0 ,subset =['continent'],inplace=True)
```

```python
df.isna().sum()
```

###  d. Fill all missing values with 0

```python
df.columns
```

```python
for i in [ 'total_cases','total_deaths','gdp_per_capita','human_development_index']:
    df[i].fillna(0,inplace=True)
```

```python
df.isna().sum()
```

## 6. Date time format :

### a. Convert date column in datetime format using pandas.to_datetime

```python
df.dtypes
```

```python
df['date']=pd.to_datetime(df['date'])
```

```python
df.dtypes
```

 ### b. Create new column month after extracting month data from date column.

```python
df['day']=df['date'].dt.day
```

```python
df['month']=df['date'].dt.month
```

```python
df.head()
```

## 7. Data Aggregation:
###        a. Find max value in all columns using groupby function on 'continent' column
###        Tip: use reset_index() after applying groupby

```python
df.groupby(['continent']).max()
```

```python
df.reset_index(inplace = True) 
df.head()
```

### b. Store the result in a new dataframe named 'df_groupby'.
###       (Use df_groupby dataframe for all further analysis)

```python
df_groupby = df
```

```python
df_groupby.head()
```

## 8. Feature Engineering :
### a. Create a new feature 'total_deaths_to_total_cases' by ratio of 'total_deaths' column to 'total_cases'

```python
df_groupby['total_deaths_to_total_cases'] = df_groupby['total_deaths'] / df_groupby['total_cases']
```

```python
df_groupby.head()
```
